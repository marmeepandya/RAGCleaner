```python
import pandas as pd
import os

# ── File mapping ──────────────────────────────────────────────────────────────
FILES = [
    "exp8_1_llm_only.csv",
    "exp8_1_rag.csv",
    "exp9_llm_only.csv",
    "exp9_rag.csv",
    "exp10_easy_llm_only.csv",
    "exp10_easy_rag.csv",
    "exp10_confusing_llm_only.csv",
    "exp10_confusing_rag.csv",
    "exp10_bge_rag.csv",
]

# ── Build master from first available file ────────────────────────────────────
master = None

for f in FILES:
    if not os.path.exists(f):
        print(f"   Not found: {f}")
        continue

    df = pd.read_csv(f)
    col_name = f.replace(".csv", "")

    # On first file, initialise master with index columns
    if master is None:
        master = df[["df1_idx", "attribute", "is_numeric", "ground_truth"]].copy()
        master = master.sort_values(["df1_idx", "attribute"]).reset_index(drop=True)

    # Merge predicted column with custom name
    pred = df[["df1_idx", "attribute", "predicted"]].rename(
        columns={"predicted": col_name}
    )
    master = master.merge(pred, on=["df1_idx", "attribute"], how="left")
    print(f"  ✓ Added column: {col_name}")

# ── Save ──────────────────────────────────────────────────────────────────────
master.to_csv("master_predictions.csv", index=False)
print(f"\n✓ Saved master_predictions.csv — shape: {master.shape}")
print(master.head(10).to_string())
```
