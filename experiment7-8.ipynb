```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
exp7_rag = pd.read_csv("exp7_all.csv")
exp7_rag["config"] = "RAG (clean KB)"

exp8_rag = pd.read_csv("exp8_all.csv")
exp8_rag["config"] = "RAG (noisy KB)"

exp7_llm = pd.read_csv("exp7_per_attr_llm.csv")
exp7_llm["config"] = "LLM-only (clean KB)"

exp8_llm = pd.read_csv("exp8_per_attr_llm.csv")  # same LLM-only for both
exp8_llm["config"] = "LLM-only (noisy KB)"
```

```python
exp7 = pd.concat([exp7_llm, exp7_rag], ignore_index=True)

CONFIG_ORDER = [
    "LLM-only (clean KB)", "RAG (clean KB)"
]

# Build heatmap data
heatmap_data = (
    exp7.groupby(["attribute", "config"])["correct_standard"]
    .mean()
    .unstack("config")
    .reindex(columns=[c for c in CONFIG_ORDER if c in exp7["config"].unique()])
)

print("Heatmap data:")
print(heatmap_data.round(2))

# ── Plot 1: Heatmap — per-attribute standard accuracy across configs ───────────
fig, ax = plt.subplots(figsize=(10, 5))
sns.heatmap(
    heatmap_data,
    annot=True, fmt=".2f",
    cmap="PuBu", vmin=0, vmax=1,
    linewidths=0.5, linecolor="white",
    ax=ax, cbar_kws={"label": "Accuracy"}
)
ax.set_title("Per-Attribute Accuracy: LLM vs RAG × Clean vs Noisy KB", fontsize=13, pad=12)
ax.set_xlabel("Configuration", fontsize=11)
ax.set_ylabel("Attribute", fontsize=11)
plt.xticks(rotation=15, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
# plt.savefig("fig_heatmap_all_configs.png", dpi=150)
plt.show()
```
